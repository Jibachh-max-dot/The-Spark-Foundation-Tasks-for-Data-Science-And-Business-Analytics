```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches=pd.read_csv('matches.csv')
matches.head()
```

```python
#this gives the information of datas.
matches.info()
```

```python
#this code helps to check if there is any missing values.
matches.isnull().sum()
```

```python
#these are total columns in our match dataset.
matches.columns
```

```python
# Total number of matches
matches['id'].max()
```

```python
# Total seasons on dataset
matches['season'].unique()
```

```python
#to check in which venue has most number of matches
plt.figure(figsize=(12,6))
sns.countplot(x='venue', data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#in 2013 season are the most of matches occured which is visualized by below bar graph.
plt.figure(figsize=(12,6))
sns.countplot(x='season', data=matches)
plt.show()
```

```python
from wordcloud import WordCloud
from scipy import signal
teamwon = matches.winner.dropna(False)
plt.subplots(figsize=(10,10))
wordcloud = WordCloud(background_color = 'white',
                     width = 512,
                     height = 384).generate(' '.join(teamwon))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
most_won = matches.winner.value_counts()[:10]
```

```python
#here we can clearly see the teams who won the most of matches.
fig, ax = plt.subplots(figsize=(15,8))
ax.set_ylim([0,120])
ax.set_ylabel("Count")
ax.set_title("Team that won the most matches")
most_won.plot.bar()

sns.barplot(x = most_won.index, y = most_won, orient='v', palette="Blues");
plt.show()
```

```python
from wordcloud import WordCloud
from scipy import signal
bestplayer = matches.player_of_match.dropna(False)
plt.subplots(figsize=(10,10))
wordcloud = WordCloud(background_color = 'white',
                     width = 512,
                     height = 384).generate(' '.join(bestplayer))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
top_players = matches.player_of_match.value_counts()[:10]
```

```python
#let us see who is the most player of the match.
fig, ax = plt.subplots(figsize=(15,8))
ax.set_ylim([0,25])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()

sns.barplot(x = top_players.index, y = top_players, orient='v', palette="Blues");
plt.show()
```

```python
#most number of matches won according to the season.
number_played = pd.melt(matches, id_vars=['id','season'], value_vars=['team1', 'team2'])

plt.figure(figsize=(12,6))
sns.countplot(x='value', data=number_played)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#the team won in final match of every season.
champions = matches.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].reset_index(drop=True)
champions
```

```python
#lets creates a pie chart to visualize the percentage of different toss decisions made in the matches dataset.
toss_ = matches.toss_decision.value_counts()
labels = (np.array(toss_.index))
sizes = (np.array((toss_/ toss_.sum())*100))
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', startangle=50)
plt.title("Toss decision percentage")
plt.show()
```

```python
#leta plot a bar graph to visualize the percentage of different toss decisions made in the matches in every dataset.
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='toss_decision', data=matches)
plt.xticks(rotation='45')
plt.show()
```

```python
# Since there is a very strong trend towards batting second let us see the win percentage of teams batting second.
num_of_wins = (matches.win_by_wickets>0).sum()
num_of_loss = (matches.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', startangle=90)
plt.title("Win percentage of team batting at second")
plt.show()
```
